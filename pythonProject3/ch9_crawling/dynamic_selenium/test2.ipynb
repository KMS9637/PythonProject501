```python
# pip install selenium
# 공식 문서
# https://www.selenium.dev/documentation/
# 깃허브 전체 소스 
# https://github.com/SeleniumHQ/seleniumhq.github.io/blob/trunk//examples/python/tests/hello/hello_selenium.py

from re import I
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
import urllib.request
import os

# 크롬 , 이미지를 검색하는 공간. 
# 검색할 키워드 목록 리스트, 
search_words = ["강동원" , "사과","바나나"]
# 웹브라우저 크롬 드라이버 로드 
driver = webdriver.Chrome()  #chromedriver.exe -> 가져 옴
# 이미지 경로를 추출. 
driver.get("https://www.google.co.kr/imghp?hl=ko&tab=wi&authuser=0&ogbl")
# assert "Python" in driver.title
# getcwd : 현재 작업 폴더를 의미. 
current_path = os.getcwd() # 저장 경로
# 검색어 , 예) 강동원 , 크로코 농구
for search_word in search_words:
    elem = driver.find_element(By.NAME,"q")  # 검색어의 검색엔진을 찾는 코드
    # 이미지 검색어 창을 비우기
    elem.clear()
    # 검색창에 자동으로 강동원이라고 쓰기
    elem.send_keys(search_word)  # 해당 검색엔진에 글자를 입력
    # 엔터 치는 명령어 
    elem.send_keys(Keys.RETURN)
    # 1초 간격으로 스크롤 대기시간, 
    SCROLL_PAUSE_TIME = 1

    # Get scroll height
    # 자바스크립트의 내장 객체이용. 
    last_height = driver.execute_script(
        "return document.body.scrollHeight")  #브라우져 높이를 확인 가능(자바스크립트)

    while True:
        # Scroll down to bottom
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);"
                              )  # 브라우져 끝까지 스크롤을 내리겠다.

        # Wait to load page
        time.sleep(SCROLL_PAUSE_TIME)  # 페이지 로딩 될 동안 웨잇

        # Calculate new scroll height and compare with last scroll height
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:  # 스크롤이 끝까지 내려가서 더이상 내릴 것이 없을 때
            try:
                # 스크롤을 끝까지 내렸음. 그리고, 혹시나, 더보기 같은 버튼이 있을 때
                # 클릭하라고, 버튼의 클래스의 요소를 선택해서,
                driver.find_elements(By.CSS_SELECTOR,".mye4qd").click()
            except:
                break
        last_height = new_height
        

    # 조회한 이미지들의 테이블 형식의 각 요소 클래스명 
    images = driver.find_elements(By.CSS_SELECTOR,".H8Rx8c")

# 강동원
    folder_name = search_word
    #해당 이미지 이름과 동일한 폴더 생성
    # 해당 경로에 폴더가 없다면
    if not os.path.isdir(folder_name):  # 없으면 새로 생성하는 조건문
        # 새로 만들기
        os.mkdir(folder_name)

#   임의로 사진 100장만 뽑을려고 , 상태 변수를 지정
    count = 1
    # 
    for image in images:
        if count > 100:
        # if count > len(images):
            break
        try:
            # 한장의 이미지를 클릭 , 명령어
            image.click()
            # 인터넷속도, 이미지 로드 시간이 걸려서, 3초 대기.  
            time.sleep(3)
           # 해당 경로는 수시로 바뀜. 핵심. 
#             각 이미지를 클릭시 , 해당 이미지의 경로 추출, 
# copy -> xpath 추출. 패턴을 파악. 
# 
# //*[@id="Sva75c"]/div[2]/div[2]/div/div[2]/c-wiz/div/div[3]/div[1]/a/img[1]
# 
# //*[@id="Sva75c"]/div[2]/div[2]/div/div[2]/c-wiz/div/div[3]/div[1]/a/img[1]
# 
# //*[@id="Sva75c"]/div[2]/div[2]/div/div[2]/c-wiz/div/div[3]/div[1]/a/img[1]
            imgUrl =  driver.find_element(By.XPATH,'//*[@id="Sva75c"]/div[2]/div[2]/div/div[2]/c-wiz/div/div[3]/div[1]/a/img[1]').get_attribute("src")
            #이미지 다운로드, 
            # 해당 경로의 이미지 url 주소에서, 현재 내가 지정한 경로에 다른이름으로 저장.
            urllib.request.urlretrieve(
                imgUrl,
                folder_name + "/" + search_word + "." + str(count) + ".jpg")
            count = count + 1
        except:
            pass
    # 해당 드라이버의 처음으로 돌아가기.    
    driver.back()

driver.close()
```
